```python
def weight(x_i, X, tau):
    '''
    Calculate weight at the x_i point
    '''
    delta_X = x_i - X
    power = -delta_X * delta_X / (2 * tau**2)
    w = np.exp(power)
    W = np.diag(w)
    return W
```

```python
def h_LWLR(X, y, tau, x):
    '''
        Hypothesis function of LWLR
    '''
    W = weight(x, X, tau)
    X = np.insert(X.reshape((-1, 1)), 0, 1, axis=1)
    theta = np.linalg.inv(X.T @ W @ X) @ X.T @ W @ y
    x_i = np.array([1, x]).T
    return theta.T.dot(x_i)
```

```python
def find_y_pred(x_test, X, y, tau):
    '''
        Calculat y predict from X set
    '''
    y_pred = []
    for x_query in x_test:
        y_i = h_LWLR(X, y, tau, x_query)
        y_pred.append(y_i)
    return y_pred
```

```python
import matplotlib.pyplot as plt

tau = 1
x_test = np.linspace(0, 10, 100)
y_pred = find_y_pred(x_test, X, y, tau)
print(y_pred[:5])

# Plot data
plt.plot(X, y, 'ro')
plt.plot(x_test, y_pred, 'g-')
plt.title('LWLR with tau=%s' % tau)
plt.show()
```
